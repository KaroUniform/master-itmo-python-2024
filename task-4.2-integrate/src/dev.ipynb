```python
from threadpool import integrate
from processpool import integrate as p_integrate
import math
import multiprocessing
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
import os
```

```python
cpu = multiprocessing.cpu_count()
```

```python
def test_function(x):
    return math.cos(x)
```

```python
results = []
for i in tqdm(range(1, cpu*2)):
    results.append(integrate(test_function, 0, math.pi/2, n_jobs=i))
```

```python
x = [item[1] for item in results]


plt.figure(figsize=(20, 6))
plt.plot(x, 'o-')
plt.xlabel('n_jobs')
plt.ylabel('time')
plt.title('Сравнение времени исполнения ThreadPoolExecutor в зависимости от числа воркеров')
plt.grid(True)
plt.show()
```
